<h1>solving problems in the data</h1>

```python
sales_data.info()
```

<h3>2. the missing values problem</h3>

lets fist see the index of the missing data

```python
sales_data[sales_data['price'].isnull()].index.tolist()
```

```python
sales_data[sales_data['quantity'].isnull()].index.tolist()
```

```python
sales_data[sales_data['revenue'].isnull()].index.tolist()
```

as we can opserve missing values are in different positions and we now that <MATH>revenue=quantity * price</MATH><br>
we now can derive every missing value from the other two

```python
sales_data['price'].iloc[193] =sales_data['revenue'].iloc[193] / sales_data['quantity'].iloc[193]
sales_data['price'].iloc[320] =sales_data['revenue'].iloc[320] / sales_data['quantity'].iloc[320]
sales_data['quantity'].iloc[122] = sales_data['revenue'].iloc[122] / sales_data['price'].iloc[122]
sales_data['revenue'].iloc[96] = sales_data['quantity'].iloc[96] * sales_data['price'].iloc[96]
```

we will just ignore this warning an continue our work and see if it worked

```python
sales_data.isnull().sum()
```

<h1>now after the data is cleaned lets start the real work</h1>

<h1>Answering Questions</h1>

<h3>1. What was the total revenue generated by the company over the course of the year?</h3>

```python
total_revenue = sales_data["revenue"].sum()
total_revenue
```

by knowing the agerage revenue we can compare it to the previous years to get more insights

<h3>2. Which product had the highest revenue? How much revenue did it generate?</h3>
<h4>solution 1 the item with highest revenue</h4>

```python
highest_revenue_index = sales_data["revenue"].argmax()
highest_revenue_product = sales_data["product"].iloc[highest_revenue_index]
highest_revenue = sales_data["revenue"].iloc[highest_revenue_index]
```

```python
highest_revenue_product, highest_revenue
```

<h4>solution 1 the item with highest total revenue</h4>

```python
highest_total_revenue_product = sales_data.groupby('product').agg(Sum=('revenue', np.sum))
```

```python
highest_total_revenue_product.loc[highest_total_revenue_product.idxmax()]
```

hence we know for sure that smart phones gets us more profit more than the other product which means that we might have to increase samrt phone sales

<h3>3. What was the average price of a product sold by the company?</h3>

```python
average_prices_per_product = sales_data.groupby('product').agg(average_price=('price', np.mean)).sort_values(by='average_price', ascending=False).reset_index()
average_prices_per_product
```

```python
px.pie(data_frame=average_prices_per_product, names='product', values='average_price', title="average price for each product", labels={
    'product':'product',
    'average_price':'average_price in usd' 
})
```

hence we see that price of laptops is the highest however smart phone gets us more profit so we might consider decrease the amount of laptops we sale

<h3>4. What was the total quantity of products sold by the company?</h3>

```python
total_quantity_of_products = sales_data.groupby('product').agg(quantity=('quantity', np.sum))['quantity'].sum()
total_quantity_of_products
```

looking to the total_quantity_of_products we can see the company sales for the whole year

<h3>5. Which category had the highest revenue? How much revenue did it generate?</h3>
<h4>solution 1 category that had highest revenue once</h4>

```python
###the highest revenue index had already been calculated
highest_revenue_category = sales_data["category"].iloc[highest_revenue_index]
highest_revenue_category, highest_revenue
```

<h4>solution 2 category that had the highest total revenue</h4>

```python
highest_total_revenue_category = sales_data.groupby('category').agg(total_revenue=('revenue', np.sum))
highest_total_revenue_category.loc[highest_total_revenue_category.idxmax()]
```

we know surely know that electronics gets us the highest revenue which also pushs us towards saling more electronics

<h3>6. What was the average revenue per sale?</h3>

which is basicaly the average revenue over the user

```python
average_revenue_per_sale = sales_data["revenue"].mean()
average_revenue_per_sale
```

by knowing the average revenue we can compare it the previous years to have more insights

<h3>7. What was the total revenue generated in each quarter of the year? (i.e. Q1, Q2, Q3, Q4)</h3>

```python
####first Dividing year into quarters
sales_data["quarters"] = sales_data['date'].dt.to_period('Q').dt.strftime('q%q')
```

```python
sales_data.head(3)
```

<h4>now after it worked we can aggregate it to find revenue at each quarter</h4>

```python
revenue_at_quarter = sales_data.groupby('quarters').agg(total_revenue=('revenue', np.sum)).reset_index()
revenue_at_quarter
```

```python
px.pie(data_frame=revenue_at_quarter, names='quarters', values='total_revenue', labels={
    'quarters':'quarters',
    'total_revenue':'total_revenue in usd'
}, title="total revenue for each year quarter")
```

by looking to this chart we can opserve that the total revenue of each quarter of the year is nearly the same, which means that our sales don't affect by much with seasons, which is basicaly amazing
